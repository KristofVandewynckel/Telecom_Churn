```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset.csv')
pd.set_option('display.max_columns', None)
df.columns = df.columns.str.lower()
df.head(20)
```

```python
df.columns
```

```python
for idx, row in df.iterrows():
    if df.loc[idx,'phoneservice'] == 'No':
        df.loc[idx,'multiplelines'] = 'No'
    if df.loc[idx,'internetservice'] == 'No':
        df.loc[idx,['onlinesecurity','onlinebackup','deviceprotection','techsupport','streamingtv','streamingmovies']] = 'No'
```

```python
df.head(20)
```

```python
print('Cust. ID: ',df['customerid'].value_counts())

print('Gender: ',df['gender'].value_counts())

print('Senior: ',df['seniorcitizen'].value_counts())

print('Partner: ',df['partner'].value_counts())

print('Dependents: ',df['dependents'].value_counts())

print('# months: ',df['tenure'].value_counts())

print('Phone service: ',df['phoneservice'].value_counts())

print('Multiple Lines: ',df['multiplelines'].value_counts())

print('Internet: ',df['internetservice'].value_counts())

print('Online security: ',df['onlinesecurity'].value_counts())

print('Online backup: ',df['onlinebackup'].value_counts())

print('Device Protection: ',df['deviceprotection'].value_counts())

print('Tech Support: ',df['techsupport'].value_counts())

print('Streaming tv: ',df['stremingtv'].value_counts())

print('Streaming movies: ',df['streamingmovies'].value_counts())

print('contract: ',df['contract'].value_counts())

print('Paperless: ',df['paperlessbilling'].value_counts())

print('Payment method: ',df['paymentmethod'].value_counts())

print('Monthly charge: ',df['monthlycharges'].value_counts())

print('DTotal charge: ',df['totalcharges'].value_counts())

print('Churn?: ',df['churn'].value_counts())
```
