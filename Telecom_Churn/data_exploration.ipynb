```python
# Import the libraries 
import pandas as pd
import numpy as np
import seaborn as sns # For creating plots
import matplotlib.ticker as mtick # For specifying the axes tick format 
import matplotlib.pyplot as plt
```

```python
# Read the CSV file 
df = pd.read_csv("telco.csv")
df.head(7)
```

```python
# Making it display all columns in the table 
with pd.option_context('display.max_rows', None):
    display(df.head(10))
```

```python
# Making the names of the columns lowercase to make it easier
df.columns = df.columns.str.lower()
```

```python
# Checking the types of the data we have 
df.dtypes
```

```python
df.info()
```

```python
# Iterating through all the columns to understand the contents of the data
for each in df.columns:
    print(f"Number of {each}: {df[each].nunique()}")
    print(f"Options of {each}: {df[each].unique()}")
    print(f"Value counts: {df[each].value_counts()}")
    print()
```

```python
# Changing "No..something" to "No" 
for idx, row in df.iterrows():
    if df.loc[idx,'phoneservice'] == 'No':
        df.loc[idx,'multiplelines'] = 'No'
    if df.loc[idx,'internetservice'] == 'No':
        df.loc[idx,['onlinesecurity','onlinebackup','deviceprotection','techsupport','streamingtv','streamingmovies']] = 'No'
```

```python
# Changing the column totalcharges from str to float
df.totalcharges = pd.to_numeric(df.totalcharges, errors='coerce')
```

```python
# Checking the number of null values
df.isnull().sum()
```

```python
# The totalcharges are NaN when tenure == 0
check = df[df["tenure"]==0] 
check.head(12)
```

```python
# Filling total charges with 0
df["totalcharges"].fillna(0, inplace = True)
```

```python
# Checking for NaN values
df.isnull().sum()
```

```python
# Dropping customerid as it brings no meanining
df_dum = df.drop(columns=["customerid"])
df_dum.head(3)
```

```python
# Looping through categorical columns 
columns=[]
for i in df_dum.columns:
    if len(df_dum[i].unique()) == 2:
        columns.append(i)
columns
```

```python
# Applying dummies on all columns 
dumdum=pd.get_dummies(df_dum,columns=columns,drop_first=True)
dumdum=pd.get_dummies(dumdum, columns =["internetservice","contract","paymentmethod",])
dumdum.head(10)
```

```python
# Saving the CSV to use it later
dumdum.to_csv("Telecom_dummies.csv", index=False)
```
